```python
import pandas as pd
```

```python
df = pd.read_csv('us_exports.csv')
```

```python
df
```

```python
years = [str(y) for y in range(2015, 2025)]
df[years] = df[years].replace(',', '', regex=True).astype(float)


# Sum export values by country (skip market share calculation)
country_totals = df.groupby('Partner')[years].sum().round(1)  # Round to 1 decimal

# Get top 10 countries by 2024 exports
top_10 = country_totals['2024'].sort_values(ascending=False).head(10).index
country_totals_top = country_totals.loc[top_10]
```

```python
country_totals_top
```

```python
share_df = (country_totals_top.div(country_totals_top.loc['Grand Total']) * 100).round(2)

# Remove the Grand Total row (now 100% everywhere)
share_df = share_df.drop('Grand Total')

# Add a "Others" row (100% - sum of shown partners)
shown_partners_sum = share_df.sum()
share_df.loc['Others'] = 100 - shown_partners_sum

share_df.loc['Grand Total (USD)'] = country_totals_top.loc['Grand Total'] 
```

```python
share_df
```

```python
years = [str(y) for y in range(2015, 2025)]
of[years] = of[years].replace(',', '', regex=True).astype(float)


# Sum export values by country (skip market share calculation)
of_totals = of.groupby('Partner')[years].sum().round(1)  # Round to 1 decimal

of_totals_top = of_totals.loc[top_10]
```
