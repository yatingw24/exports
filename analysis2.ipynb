```python
import pandas as pd
```

```python
df = pd.read_csv('FRZ.csv')
```

```python
#frozen boneless meat

df.groupby('Partner')
```

```python
years = [str(y) for y in range(2015, 2025)]
df[years] = df[years].replace(',', '', regex=True).astype(float)

frozen_pork = df[df['HS Code'].astype(str).str.startswith('2032')]

global_totals = frozen_pork[years].sum()

market_share = (
    frozen_pork.groupby('Partner')[years].sum()  # Sum exports by country
    .div(global_totals, axis='columns') * 100    # Divide by yearly global totals
)

market_share = market_share.round(1)

top_10 = market_share['2024'].sort_values(ascending=False).head(10).index
market_share_top = market_share.loc[top_10]

market_share_top
```

```python
df_of = pd.read_csv('frz_of.csv')
```

```python
years = [str(y) for y in range(2015, 2025)]
df_of[years] = df_of[years].replace(',', '', regex=True).astype(float)

frozen_of = df_of[df_of['HS Code'].astype(str).str.startswith('2064')]

global_totals_of = frozen_of[years].sum()

market_share_of = (
    frozen_of.groupby('Partner')[years].sum()  
    .div(global_totals_of, axis='columns') * 100   
)

market_share_of = market_share_of.round(1)

top_10_of = market_share_of['2024'].sort_values(ascending=False).head(10).index
market_share_top_of = market_share_of.loc[top_10]

market_share_top_of
```
